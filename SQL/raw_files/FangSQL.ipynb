```python

class ScipySEMAggregate(object):
    def __init__(self):
        self.values = []
    def step(self, value):
        self.values.append(int(value))
    def finalize(self):
        return stats.sem(self.values)
con.create_aggregate("sp_sem", 1, ScipySEMAggregate)
#cursor.execute('''SELECT CURRENTGRADE FROM WebExtract_f
#                    WHERE CURRENTGRADE>0;''')
#a = cursor.fetchall()
#print a

cursor.execute('''SELECT ZIPCODE, AVG(SCORE) AS MEANGRADE, 
                    sp_sem(SCORE) AS SEGRADE, COUNT(*) AS NUM_INSP FROM WebExtract_f
                    WHERE SCORE>0 AND zipcode <> 'N/A' AND cast(zipcode as int) >0 
                    GROUP BY ZIPCODE
                    ORDER BY MEANGRADE
                    ;''')
#cursor.execute("SELECT * FROM WebExtract_f;")
a = cursor.fetchall()
print a
```
